```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
from datetime import datetime as dt
data=[]
times=[]
with open("test_02#04#2019#21#14#10.txt", "r") as f:
    data=eval(f.readline())
    times=eval(f.readline())
    rez_coef=eval(f.readline())
times=[dt.strptime(s,"%d#%m#%Y#%H#%M#%S#%f") for s in times]
```

```python
from struct import *
rezs=[]
for s in data:
    fmt="h"*(len(s)//2)
    rezs.append(unpack(fmt,s.encode()))
```

```python
rezx=[(x-times[0]).total_seconds() for x in times]
mins=[min(x)/rez_coef for x in rezs]
maxs=[max(x)/rez_coef for x in rezs]
delta=[maxs[i]-mins[i] for i in range(len(mins))]
```

```python
trace0 = go.Scatter(
    x=rezx,
    y=mins,
    name='min'
)
trace1 = go.Scatter(
    x=rezx,
    y=maxs,
    name='max'
)
trace2 = go.Scatter(
    x=rezx,
    y=delta,
    name='delta'
)
# определяем массив данных и задаем title графика в layout
data = [trace0,trace1,trace2]
layout = {'title': 'Measurements'}

# cоздаем объект Figure и визуализируем его
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```
