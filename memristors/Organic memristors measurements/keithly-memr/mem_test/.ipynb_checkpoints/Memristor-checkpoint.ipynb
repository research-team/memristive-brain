```python
# отключим предупреждения Anaconda
import warnings
warnings.simplefilter('ignore')

# будем отображать графики прямо в jupyter'e
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
#графики в svg выглядят более четкими
%config InlineBackend.figure_format = 'svg' 

#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 8, 5
import pandas as pd
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
df = pd.read_csv('defbuffer1.csv')
df.head()
```

```python
trace0 = go.Scatter(
    x=df['Value'],
    y=df['Reading'],
    name='IV'
)

# определяем массив данных и задаем title графика в layout
data = [trace0]
layout = {'title': 'IV'}

# cоздаем объект Figure и визуализируем его
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```

```python
R=df['Value']/df['Reading']
R.head()
```

```python
trace0 = go.Scatter(
    x=list(range(0,len(R))),
    y=R,
    name='R'
)

# определяем массив данных и задаем title графика в layout
data = [trace0]
layout = {'title': 'R'}

# cоздаем объект Figure и визуализируем его
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```

```python
mem1u = pd.read_csv("C:\\Users\\itis\\Documents\\Anaconda\\mem_test\\mem1up.csv")
mem1d = pd.read_csv("C:\\Users\\itis\\Documents\\Anaconda\\mem_test\\mem1down.csv")
Rmem1u=pd.Series(mem1u['Value']/mem1u['Reading'])
Rmem1d=pd.Series(mem1d['Value']/mem1d['Reading'])
```

```python
Rmem1u=Rmem1u[Rmem1u>100000][Rmem1u<600000]
Rmem1d=Rmem1d[Rmem1d>100000][Rmem1d<350000]
```

```python
trace0 = go.Scatter(
    x=mem1u["Value"],
    y=mem1u["Reading"],
    name='mem1u'
)
trace1 = go.Scatter(
    x=mem1d["Value"],
    y=mem1d["Reading"],
    name='mem1d'
)

# определяем массив данных и задаем title графика в layout
data = [trace0, trace1]
layout = {'title': 'mem1IV'}

# cоздаем объект Figure и визуализируем его
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```

```python
trace0 = go.Scatter(
    x=list(range(0,len(Rmem1u))),
    y=Rmem1u,
    name='Rmem1u'
)
trace1 = go.Scatter(
    x=list(range(len(Rmem1d),0,-1)),
    y=Rmem1d,
    name='Rmem1d'
)

# определяем массив данных и задаем title графика в layout
data = [trace0,trace1]
layout = {'title': 'mem1R'}

# cоздаем объект Figure и визуализируем его
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```
